```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df_walmart3['IsHoliday']= label_encoder.fit_transform(df_walmart3['IsHoliday'])
df_walmart3['Type']= label_encoder.fit_transform(df_walmart3['Type'])
```

```python
df_walmart3
```

```python
df_walmart4=df_walmart3.drop(['Date'],axis=1)
df_walmart4
```

```python

from sklearn.ensemble import RandomForestRegressor
```

```python
no_weeksales=df_walmart4.drop(['Weekly_Sales'],axis=1)
weeksales=df_walmart4['Weekly_Sales']
```

```python
weeksales
```

```python
rndf = RandomForestRegressor(n_estimators=25)
rndf.fit(no_weeksales,weeksales)
```

```python

plt.barh(no_weeksales.columns, rndf.feature_importances_)
```
